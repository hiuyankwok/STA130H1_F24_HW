2. continuous variable vs indicator variable in SLR
continuous variable:  take any value within a range and represents measurements or quantities (e.g., height, weight).
Linear form with continuous variable: outcome = B0 + B1 x continuous predictor

Indicator variable:  takes the value of 1 or 0 to represent the presence or absence of a categorical attribute (e.g., gender, treatment vs. control).
Linear form with continuous variable: outcome = B0 + B1 x 1 (indicator)

3. Behaviour change when indicator variable(s) are added in multiple lienar regression
When a Multiple Linear Regression model includes both a continuous variable and an indicator variable, the model will estimate different intercepts for the groups defined by the indicator variable, while maintaining the same slope for the continuous predictor across those groups. the indicator added , allows the equation for an adjusted intercept for different groups.

outcome = B0 + B1 x continuous predictor + B2x 1(indicator)

4. interaction between continuous and indicator variable in MLR
an interaction term between a continuous and an indicator variable in the model allows the slope of the continuous variable to differ across the levels of the indicator variable. The slope for the continuous predictor differs depending on the category represented by the indicator variable.

outcome = B0 + B1 x continuous predictor + B2x 1(indicator) + B3 x (contunuous predictor x 1(indicator)

5. Multiple Linear Regression with indicator variables for a NON-BINARY categorical variable

when a non-binary categorical variable is added into MLR, it must be encoded using indicator variables. If the categorical variable has

Linear form: B0 + B1 x 1(category1) + B2 x 1(category2) + ... + B(c-1) x 1(category(c-1))

> _The way an **interaction** actually appears and works in the context of the linear form of a **multiple linear regression** model may not necessarily be immediately intuitive, as it is actually in fact somewhat subtle and tricky. Of course, an **interaction** is when the relationship of one **predictor variable** with the **outcome variable** depends on the value of another different **predictor variable**, so the impact of one **predictor variable** changes based on the presence or magnitude of another **predictor variable**. But are you sure you know what this means in the context of the **linear form** of a **multiple linear regression** model?_
>
> - _Imagine you're selling smoothies, the taste of your smoothie depends on the ingredients included in the smoothie, and there are two ingredients (bananas and strawberries) in the smoothie._
>     - _Adding more bananas into the smoothie will of course increase the "banana" flavor of the smoothie, and vice-versa for strawberries..._
>     - _But is this "banana" influence on the flavor always constant, or can it change depending on the absolute amount of strawberries in the smoothie?_ 
>     - _If the "banana" flavor influence is constant and does not depend on the  absolute amount of strawberries in the smoothie, then there is no **interaction** and the **linear form** of the model is $\beta_b b_i + \beta_s s_i$ and the model is said to be only **additive**._
>     - _But if the "banana" flavor influence does depend on the absolute amount of strawberries in the smoothie, then there IS an **interaction** and the **linear form** of the model is $\beta_b b_i + \beta_s s_i + \beta_{bs} (b_i\times s_i)$ and the model is said to be **synergistic**._
>         
> _These **linear forms** show that either bananas and strawberries do not have any **synergistic interaction** and contribute to the flavor independently; or, they do have a **synergistic interaction** and there is an interesting interplay between bananas and strawberries in the way they influence the taste of the smoothie._ 
> 
> - _So, if there is no **interaction**, then the effect of adding more bananas on the taste of the smoothie will always be the same, no matter how many strawberries you put in. So the effect of bananas on the smoothie is the same whether you add a lot of strawberries or just a few: $\beta_b b_i + \beta_s s_i$_
> - _Or, on the other hand, if there is an **interaction**, then the effect of adding bananas (on the smootie flavor) will be different depending on how many strawberries there currently are in the smoothie: $\beta_b b_i + \beta_s s_i + \beta_{bs} (b_i\times s_i)$_
> 
> _In this case, the right answer is probably that the **linear form** with the **interaction** is correct. This is because the flavor probably depends on the relative amount of bananas and strawberries in the smoothie; so, the effect of adding a fixed amount of bananas to the smoothie probalby depends on the absolute amount of strawberries that are in the smoothie._
> 
> _Again, because understanding **interactions** in the context of **linear forms** is somewhat subtle and tricky and indeed not necessarily obviously intuitive, let's think about this a bit more.
````python
import pandas as pd
import statsmodels.formula.api as smf
import plotly.graph_objects as go
import numpy as np

# Load the dataset
url = "https://raw.githubusercontent.com/pointOfive/stat130chat130/main/CP/CSCS_data_anon.csv"
cscs_data = pd.read_csv(url)

# Prepare the data: Convert categorical variables into binary ones for logistic regression
# Example: Convert group_exercise to binary (participated or not)
cscs_data['group_exercise'] = cscs_data['CONNECTION_activities_group_exercise_p3m'].apply(lambda x: 1 if x in ['Weekly', 'Daily or almost daily'] else 0)

# Example: Convert preference_time_friends to binary (more than 1 hour of socializing or not)
cscs_data['preference_time_friends'] = cscs_data['CONNECTION_preference_time_friends_grouped'].apply(lambda x: 1 if x in ['5 or more hours', '1 to 4 hours'] else 0)

# Outcome variable: WELLNESS_phq_feeling_down (convert it to binary: feeling down vs not feeling down)
cscs_data['WELLNESS_phq_feeling_down_binary'] = cscs_data['WELLNESS_phq_feeling_down'].apply(lambda x: 1 if x in ['Nearly every day', 'More than half the days', 'Several days'] else 0)

# Define the logistic regression model formula
formula = 'WELLNESS_phq_feeling_down_binary ~ group_exercise * preference_time_friends'

# Fit the logistic regression model
log_reg_fit = smf.logit(formula, data=cscs_data).fit()

# Display model summary
print(log_reg_fit.summary())

# For visualization: Simulate data to plot best fit lines
# Generate a range of values for the predictors (group_exercise and preference_time_friends)
group_exercise_vals = np.linspace(0, 1, 100)
preference_time_vals = np.linspace(0, 1, 100)

# Create a meshgrid for the interaction plot
X, Y = np.meshgrid(group_exercise_vals, preference_time_vals)

# Calculate the log-odds using the model's coefficients
log_odds = log_reg_fit.params['Intercept'] + log_reg_fit.params['group_exercise'] * X + log_reg_fit.params['preference_time_friends'] * Y + log_reg_fit.params['group_exercise:preference_time_friends'] * X * Y

# Convert log-odds to probabilities (using the logistic function)
probabilities = 1 / (1 + np.exp(-log_odds))

# Plot the 3D surface plot of the probabilities
fig = go.Figure(data=[go.Surface(z=probabilities, x=X, y=Y, colorscale='Viridis')])

# Labels for axes
fig.update_layout(title='Probability of Feeling Down vs Group Exercise & Socializing',
                  scene=dict(xaxis_title='Group Exercise', yaxis_title='Socializing Time', zaxis_title='Probability of Feeling Down'))

# Show the plot
fig.show()
````

Goal of the Demonstration:

1. assess the stability and generalizability of the model by performing multiple random splits of the data, fitting the model each time, and evaluating its performance both on the training set ("in-sample") and the testing set ("out-of-sample"). 
2. evaluate if the model is overfitting or underfitting. 
3. generalization: This approach helps to understand how well the model generalizes to new data. The primary goal is not just to fit the training data well but to ensure that the model can perform accurately on new, unseen data.


Repeatedly running this experiment -> helps understand how the model behaves across different data splits

steps of the demo:
1. Random Splitting: 
We'll repeatedly split the dataset into training and test sets (without fixing the random seed) to simulate different data conditions.

2. Model Training: 
Each time, we'll fit a model using the training data.

3. Model Evaluation: 
We’ll calculate both in-sample and out-of-sample R-squared values to assess the model’s fit and its ability to generalize.

4. Visualization: 
We will visualize these values to identify any patterns, like overfitting or underfitting.

````python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf

# Assuming 'pokeaman' dataset is loaded and processed
# Replace NaN in the 'Type 2' column with 'None'
pokeaman.fillna('None', inplace=True)

# Split the dataset into training and testing (50-50 split)
fifty_fifty_split_size = int(pokeaman.shape[0] * 0.5)
pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=fifty_fifty_split_size)

# Convert categorical columns in the training data
pokeaman_train['Type 1'] = pd.Categorical(pokeaman_train['Type 1'])
pokeaman_train['Type 2'] = pd.Categorical(pokeaman_train['Type 2'])

# Apply the same categories to the test set, ensuring consistency
pokeaman_test['Type 1'] = pokeaman_test['Type 1'].astype(pd.CategoricalDtype(categories=pokeaman_train['Type 1'].cat.categories))
pokeaman_test['Type 2'] = pokeaman_test['Type 2'].astype(pd.CategoricalDtype(categories=pokeaman_train['Type 2'].cat.categories))

# Define the linear form for the model
model_linear_form = 'HP ~ Attack + Defense + Speed + Legendary + Q("Sp. Def") + Q("Sp. Atk") + C(Generation) + C(Q("Type 1")) + C(Q("Type 2"))'

# Set up the loop for collecting the performance metrics
reps = 100
in_sample_Rsquared = np.array([0.0] * reps)
out_of_sample_Rsquared = np.array([0.0] * reps)

for i in range(reps):
    # Split the data again without setting a fixed random seed
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=fifty_fifty_split_size)
    
    # Convert categorical columns in the new training data
    pokeaman_train['Type 1'] = pd.Categorical(pokeaman_train['Type 1'])
    pokeaman_train['Type 2'] = pd.Categorical(pokeaman_train['Type 2'])
    
    # Apply the same categories to the test set
    pokeaman_test['Type 1'] = pokeaman_test['Type 1'].astype(pd.CategoricalDtype(categories=pokeaman_train['Type 1'].cat.categories))
    pokeaman_test['Type 2'] = pokeaman_test['Type 2'].astype(pd.CategoricalDtype(categories=pokeaman_train['Type 2'].cat.categories))
    
    # Fit the model
    model_spec = smf.ols(formula=model_linear_form, data=pokeaman_train)
    model_fit = model_spec.fit()
    
    # Store the in-sample R-squared value
    in_sample_Rsquared[i] = model_fit.rsquared
    
    # Get predictions on the test set and calculate out-of-sample R-squared
    yhat = model_fit.predict(pokeaman_test)
    y_test = pokeaman_test['HP']
    out_of_sample_Rsquared[i] = np.corrcoef(y_test, yhat)[0, 1] ** 2

# Create a DataFrame to store the results
df = pd.DataFrame({
    "In Sample Performance (Rsquared)": in_sample_Rsquared,
    "Out of Sample Performance (Rsquared)": out_of_sample_Rsquared
})

# Visualize the results
import plotly.express as px
import plotly.graph_objects as go

# Scatter plot for visualization
fig = px.scatter(df, x="In Sample Performance (Rsquared)", y="Out of Sample Performance (Rsquared)")
fig.add_trace(go.Scatter(x=[0, 1], y=[0, 1], name="y=x", line_shape='linear'))

fig.show()
````

````python
import matplotlib.pyplot as plt

plt.scatter(df["In Sample Performance (Rsquared)"], df["Out of Sample Performance (Rsquared)"])
plt.plot([0, 1], [0, 1], linestyle='--', label="y=x")
plt.xlabel('In Sample Rsquared')
plt.ylabel('Out of Sample Rsquared')
plt.legend()
plt.show()
````

Interpretting results:

- Consistency: 
If the in-sample and out-of-sample R-squared values are consistently close to each other, the model is likely well-calibrated and generalizes well to new data. This is a sign of stability in model performance.

- Overfitting: 
If the in-sample R-squared is significantly higher than the out-of-sample R-squared, it suggests that the model may be overfitting. In this case, the model performs well on the training data but fails to generalize to the test data.

- Underfitting: 
If both in-sample and out-of-sample R-squared are low, it suggests that the model is too simple and is underfitting the data, failing to capture important patterns in the data.

- Outliers or Variability: 
If the plot shows a large spread of points, it could suggest that the model is unstable, and the performance is highly sensitive to the random splits. This could indicate the need for a more robust model or more data.
