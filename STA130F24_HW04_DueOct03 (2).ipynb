Standard deviation:
- measures the average spread or dispersion of data points from the mean of the dataset. it is given by the formula of the square root of the sum of the differences of each data point and the mean (squared) divided by the number of data points.

Method 1: using standard error

as described in the video, generate bootstrapped samples of 1000s or more. calculate the its mean, and repeat this whole process multiple times, so you have a distribution of multiple sample means. Now calculate the standard deviation of the distrbution of the bootstrapped means(= standard error). Given that this distribution of means is bell-shaped:

The critical Z-value is 1.96 for a 95% confidence interval (this Z-value corresponds to the point on the standard normal curve where 2.5% of data lies in lower tail and 2.5% lies in upper tail. the smaller the Z value, the smaller the confidence interval.)

The formula for the confidence interval using the standard error is:
CI = (x - Z-value x SE, x + Z-value x SE)
where x = mean of boostrapped sample means
SE = standard error calculated from bootstrapped sample means

````python
import pandas as pd
import numpy as np

# Step 1: Load the dataset
url = 'https://raw.githubusercontent.com/Opensourcefordatascience/Data-sets/master/blood_pressure.csv'
data = pd.read_csv(url)

# Display the first few rows and check column names
print(data.head())
print(data.columns)

# Step 2: Select the relevant column for analysis
# From inspection, the column names are 'bp_before' and 'bp_after'
# Let's calculate the bootstrap confidence interval for the 'bp_before' values.
blood_pressure_before = data['bp_before']  # Column for blood pressure before treatment

# Step 3: Define the bootstrapping function
def bootstrap(data, num_samples=1000, statistic_func=np.mean):
    boot_statistics = []
    n = len(data)
    
    for _ in range(num_samples):
        # Sample with replacement
        sample = np.random.choice(data, size=n, replace=True)
        # Apply the chosen statistic function to the sample
        boot_statistics.append(statistic_func(sample))
    
    return np.array(boot_statistics)

# Step 4: Perform bootstrapping for the mean
num_samples = 1000  # Number of bootstrap samples
boot_means = bootstrap(blood_pressure_before, num_samples, statistic_func=np.mean)

# Step 5: Calculate the 95% confidence interval from bootstrapped means
ci_lower = np.percentile(boot_means, 2.5)  # 2.5th percentile
ci_upper = np.percentile(boot_means, 97.5)  # 97.5th percentile

# Step 6: Print the results
print(f"95% Confidence Interval for the Mean Blood Pressure (before): ({ci_lower}, {ci_upper})")

# Step 7: How to modify the code to calculate a confidence interval for a different population parameter
# For example, to calculate the confidence interval for the median:
boot_medians = bootstrap(blood_pressure_before, num_samples, statistic_func=np.median)

# Calculate the 95% confidence interval for the median
ci_lower_median = np.percentile(boot_medians, 2.5)
ci_upper_median = np.percentile(boot_medians, 97.5)

# Print the results for median
print(f"95% Confidence Interval for the Median Blood Pressure (before): ({ci_lower_median}, {ci_upper_median})")
````

Population parameter: the actual, fixed value that we are trying to estimate through sample statistics and confidence intervals. Eg. mean, median etc
    
Sample statistic: A sample parameter you calculate sampling a subset of from the actual data set. Thus, everytime you draw new samples, the value can value, and it is not fixed. In other words, it is subject to sampling variability.

Confidence intervals are based on a bunch of sample statistics because you don't know but are trying to find a best estimate of the population parameter. We need to distinguish between the roles of population parameter and sample statistic because the confidence interval describes where the population parameter is LIKELY to be, but NOT where the statistic statistic falls. 1 & 2. bootstrapping is a resampling technique, a process that allows us to assess the variability/ confidence intervals (useful for when you want to estimate the true parameters of a data set). It is particularly useful for when you have a small sample size or unevenly distributed dataset, and bootstrapping allows you to assess and make assumptions given the restrictions of classical statistical techniques that would otherwise cost a lot or be difficult to practise. The process is about drawing samples from a population by resampling with replacement from the original data set, calculating a certain population parameter for each sample, and using all the collected sample statistics to create some form of visualization/ analysis, such as KDE or histograms. you can then create confidence intervals based on the sample statistics (to estimate where the true parameter is), as well as calculate the standard error (which indicates how much sample statistics would fluctuate if you repeatedly sampled from the data population). There are many other things you can do with bootstrapped data as well. 

3. hypothesis testing: 
1. given that you have a hypothesized guess about the average (your null hypothesis)
2. use bootstrapping to compute the sample mean from your sample population. This sample mean is an ESTIMATE of the POPULATION mean. 
3. adjust ALL your data points in the sample BY THE SAME AMOUNT so that the sample mean match up with your hypothesized guess. This shifted data has the same variability as your original population. 
4. bootstrap your shifted data, and from each resample, calculate a new bootstrap sample mean; repeat 1000s of times. this creates a distribution of sample means UNDER THE ASSUMPTION that the null hypothesis is TRUE.
5. Calculate the bootstrapped means. This shows what sample means you mgiht expect if the population is really the null hypothesis.
6. compare the actual sample mean to the bootstrapped distribution; if it falls far out in the tails of the bootstrap distribution, then your null hypothesis is unlikely true.

The confidence interval, which can be derived using the bootstrapping process as described in the above answers, provides an estimate of a range of which we can expect the REAL POPULATION PARAMETER to fall. Quantitative analysis:
Below is the code for the bootstrapped data of both initial and final health scores, as well as the confidence interval for both respectively. The steps of bootstrapping can be found commented along its code counterpart. Since the sample size is quite small and it would be inaccurate if you use the sample mean to give estimates for the effects on other patients/ the general public, use bootstrapping to estimate the distribution of a sample statistic (e.g., mean, median, variance) by resampling the original data with replacement. This allows us to make inferences about a population even when the sample size is small or when we cannot make assumptions about the underlying distribution of the data. 

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

# Load the CSV file into a DataFrame
file_path = 'https://drive.google.com/uc?id=1ahREfTB6LJcbdLxCDxLOsMSR08uzFnOz'
data = pd.read_csv(file_path)

# Display the first few rows of the DataFrame to understand its structure
print(data.head())

# Step 1: Set the random seed for reproducibility
np.random.seed(42)  # Setting the seed for reproducibility

# Step 2: Define a function to perform bootstrapping
def bootstrap(data, num_samples=1000):
    bootstrapped_means = []
    for _ in range(num_samples):
        sample = np.random.choice(data, size=len(data), replace=True)
        bootstrapped_means.append(np.mean(sample))
    return bootstrapped_means

# Step 3: Bootstrap the means of Initial and Final Health Scores
initial_scores = data['InitialHealthScore']
final_scores = data['FinalHealthScore']

# Perform bootstrapping
bootstrapped_initial_means = bootstrap(initial_scores, num_samples=1000)
bootstrapped_final_means = bootstrap(final_scores, num_samples=1000)

# Step 4: Calculate the 95% confidence intervals
def confidence_interval(data, confidence_level=0.95):
    lower_bound = np.percentile(data, (1 - confidence_level) / 2 * 100)
    upper_bound = np.percentile(data, (confidence_level + (1 - confidence_level) / 2) * 100)
    return lower_bound, upper_bound

initial_ci = confidence_interval(bootstrapped_initial_means)
final_ci = confidence_interval(bootstrapped_final_means)

print(f"Initial Health Score CI: {initial_ci}")
print(f"Final Health Score CI: {final_ci}")

# Step 5: Visualize the distributions of bootstrapped means
plt.figure(figsize=(14, 6))

# KDE for bootstrapped initial health scores
plt.subplot(1, 2, 1)
sns.kdeplot(bootstrapped_initial_means, fill=True, color='blue', label='Bootstrapped Initial Means')
plt.title('Bootstrapped Initial Health Score Means')
plt.xlabel('Mean Health Score')
plt.ylabel('Density')
plt.axvline(x=np.mean(initial_scores), color='red', linestyle='--', label='Observed Mean')
plt.legend()

# KDE for bootstrapped final health scores
plt.subplot(1, 2, 2)
sns.kdeplot(bootstrapped_final_means, fill=True, color='orange', label='Bootstrapped Final Means')
plt.title('Bootstrapped Final Health Score Means')
plt.xlabel('Mean Health Score')
plt.ylabel('Density')
plt.axvline(x=np.mean(final_scores), color='red', linestyle='--', label='Observed Mean')
plt.legend()

plt.tight_layout()
plt.show()

# Step 6: Perform a statistical test to compare the means
t_statistic, p_value = stats.ttest_ind(initial_scores, final_scores)
print(f"T-statistic: {t_statistic}, P-value: {p_value}")

# Interpretation of Results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is evidence that the vaccine is effective.")
else:
    print("Fail to reject the null hypothesis: There is no evidence that the vaccine is effective.")
````

Conclusion

the null hypothesis that the vaccine has no effect on the patients are rejected because the entire CI for the final health scores is above the entire CI for the initial health scores, suggesting that the vaccine has a high likelihood that it raised the health scores. On average, the final health scores are higher than the initial health scores even after bootstrapping.

You can conduct power analysis to determine the min. sample size needed.
